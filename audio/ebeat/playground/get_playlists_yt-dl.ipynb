```python
from youtube_dl import YoutubeDL
```

```python
url = 'https://www.youtube.com/playlist?list=PLZkoOKH0XBibq6BJe0x7w2Yr89p7XOfVp'
ydl_opts = {
    'flat-playlist' : 'True',
    'extract_flat' : 'True',
    'skip_download' : 'True',
}
ydl = YoutubeDL(ydl_opts)
ydl.add_default_info_extractors()
playlist = ydl.extract_info(url, download=False)
```

```python
for item in playlist['entries']:
    title = item['title']
    print(title)
```
